```python
import pandas as pd
from google.colab import  drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/titanic.csv')
```

```python
#Dimensiones
print(f"Dimensiones del DataFrame : {df.shape}")
print(f"Numero de datos : {df.size}")
print(f"Nombre de columnas: {df.columns}")
print(f"Nombre de filas: {df.index}")
print(f"Tipos de datos: {df.dtypes}")
print(f"Primeras 10 filas: {df.head(10)}")
print(f"Ultimas 10 filas: {df.tail(10)}")
```

```python
#Pasadjero Id 148
print(df.loc[148])
```

```python
#Filas pares del data frame
filas_pares = df.iloc[::2]
print(filas_pares)
```

```python
#Mostrar los nombres de las personas que iban en primera clase ordenadas alfabeticamente
print(df[df["Pclass"]==1]["Name"].sort_values())
```

```python
# Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron
print(df['Survived'].value_counts()/df['Survived'].count() * 100)

# Alternativa
print(df['Survived'].value_counts(normalize=True) * 100)
```

```python
#Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase
print(df.groupby('Pclass')['Survived'].value_counts(normalize=True))
```

```python
#Eliminar del DF los pasajeros de edad desconociad y cabina desconocida
df.dropna(subset=['Cabin', "Age"])
```

```python
# Mostrar la edad media de las mujeres que viajaban en cada clase.
print(df.groupby(['Pclass','Sex'])['Age'].mean().unstack()['female'])
```

```python
# Añadir una nueva columna booleana para ver si el pasajero era menor de edad o no.
df['Young'] = df['Age'] < 18
```

```python
# Mostrar el porcentaje de menores y mayores de edad que sobrevivieron en cada clase.
print(df.groupby(['Pclass', 'Young'])['Survived'].value_counts(normalize = True) * 100)
```
