#Aplicar funciÃ³n de agregaciÃ³n por grupos
Una vez dividido el DataFrame en grupos, es posible aplicar funciones de agregaciÃ³n a cada grupo mediante el siguiente mÃ©todo


Una funcion de agregaciÃ³n toma un conjunto de datos  y devuelve un solo valore resumen


        * mean()
        
        * sum()

        * max()

        * min()

        * count()

        * median()

        * std()



Funcion agg() permite aplicar una o varias funciones a uno o varios grupos de un data frame
â MÃ¡s flexible que .mean() o .sum()
â Permite aplicar diferentes funciones a diferentes columnas
â Se ve mÃ¡s ordenado que escribir todo por separado

````python
import pandas as pd
url = 'https://raw.githubusercontent.com/asalber/manual-python/master/datos/colesterol.csv'
df = pd.read_csv(url, na_values="NaN")
df.dropna(inplace=True)
````

````python
#Promedio colesterol por sexo:

df.groupby("sexo")["colesterol"].mean()
````

````python
#Suma total de peso por sexo

df.groupby("sexo")["peso"].sum()
````

````python
#Resetear el index, para que no pase al primer parametro como indice
promedios = df.groupby("sexo")["colesterol"].mean().reset_index()
promedios
````

````python
#Varias funciones a la vez
df.groupby("sexo")["colesterol"].agg(["mean","max","min"]).reset_index()
````

````python
#Agregar multiples columnas con multiples funciones

df.groupby("sexo").agg({
    "edad": ["mean", "max"],
    "colesterol": ["mean", "min", "max"]}).reset_index()
````

#Conbinar varios DataFrames

dos o mÃ¡s DataFrames puede conbinarse en otro DataFrame. La combinacion puede ser de varias formas:
      
       * ConcatenaciÃ³n: CombinaciÃ³n de varios DataFrames concatenando sus filas o columnas

      * Mezcla: CominaciÃ³n de varios DataFrames usando columnas o Ã­ndices comunes

#ConcatenaciÃ³n de DataFrames


      ConcatenaciÃ³n de filas. Las filas de dos DataFrames se concatenan unas a continuaciÃ³n de las otras para formar el nuevo DataFrame. Para ello es necesario que los Data Frames se conbinen y tengan el mismo Ã­ndice de columnas
          Verticalmente


      concatenaciÃ³n de columnas. Las columnas de los Data Frames se concatenan unas a continuaciÃ³n de las otras para formar el nuevo DataFrame. Para ello es necesario que los DataFrame que se combinen tengan el mismo Ã­ndice de filas
          Horizontalmente




si los DataFrame que se concatenan por filas no tienen el mismo Ã­ndice de columnas, el DataFrame resultante incuirÃ¡ todas las columnas exitentes en los DataFrames y rellenarÃ¡ con valores NaN los datos no disponibles. Si los DataFrames que se concatenan por columnas no tienen el mismo Ã­ndice de filas, el DataFrame resultante incluirpa todas las filas existentes en los DataFarme y rellenarÃ¡ con valores NaN de los datos no disponibles

````python
import pandas as pd
ruta = "/content/drive/MyDrive/Bases de datos /bases_api_normalizada.xlsx"

vida =  pd.read_excel(ruta, sheet_name="esperanza_de_vida")
poblacion1 =pd.read_excel(ruta, sheet_name="poblacion")

poblacion = poblacion1[(poblacion1["id_provincia"].notna()) & (poblacion1["id_provincia"] != 25)]

print(vida.head())
print(poblacion.head())
````

#Left join

El DataFrame resultante contiene todas las filas del primer DataFrame y descarta las filas del segundo DataFrame que no puedan emparejarse con alguna fila del primer DataFrame a travÃ©s de la clave

* Se conservan todas las filas del DataFrame de la izquierda
* Y se agregan las columnas de la derecha solo si hay coincidencia
* Si no hay coincidencia, completa con un NaN


Es util cuando: Quieres mantener todos tus datos principales y agregar info complementaria si estÃ¡

````python
df_left = pd.merge(vida,poblacion, on=["id_provincia", "anio"], how="left")


print(df_left.shape) #Cuantas filas y columnas
print(df_left.head()) #Primeras fila
print(df_left.isna().sum()) #Ver si hay valores faltantes
````

#Rigth join

es exactamente lo contrario a left join:
El DataFrame resultante contiene todas las filas del segundo DataFrame y descarta las filas del primero DataFrame que no puedan emparejarse con alguna fila del segundo DataFrame a travÃ©s de la clave

* Se conservan todas las filas del DataFrame de la derecha
* Y se agregan las columnas de la izquierda solo si hay coincidencia
* Si no hay coincidencia, completa con un NaN


Es util cuando: Quieres mantener todos tus datos principales y agregar info complementaria si estÃ¡

````python
df_right = pd.merge(vida, poblacion, on=["id_provincia", "anio"], how="right")

print(df_right.shape)
print(df_right.head())
print(df_right.isna().sum())
````

#Outer join

El DataFramee resultante contiene todas las filas de los dos DataFrames. Si una fila de un DataFrane no puede emparejarse con otra los mismo valores en la clave con el otro DataFrame, la fila se aÃ±ade igualmente al DataFrame resultante rellenando las columnas del otro DataFrame con el valor NaN. Es quivalente a la uniÃ³n de conjuntos.

* Todas las filas de ambos DataFrames
* Si hay coincidencia se unen
* si no hay coincidencias, completa los datos faltantes con NaN

````python
df_outer = pd.merge(vida,poblacion, on=["id_provincia" ,"anio"], how="outer")

print(df_outer.shape)
print(df_outer.head())
print(df_outer.isna().sum())
````
